Caragamos csv y creamos dataframe

```python
import pandas as pd

df = pd.read_csv("2_air_filter.csv", encoding='utf-8-sig')
```

Creamos condicion de filtrado y creamos otro dataframe para guardar y mostrar los datos del filtrado cruzado

```python
condicionFiltro =(df["efficiency"] > 0.90) & (df["filter_age_days"] > 10)
lColumnas = ["filter_name", "filter_class", "location", "pressure_drop_pa", "replacement_needed"]
df2 = df.loc[condicionFiltro,lColumnas]
df2
```

Creamos un nuevo dataframe agrupado por clase de filtro y localización, mostrando la eficiencia media y el total de filtros que necesitan reemplazo.

```python
dfAgrupado = df.groupby(["filter_class", "location"]).agg(
    eficiencia_media = ("efficiency", "mean"),
    total_reemplazos = ("replacement_needed", "sum")
)
dfAgrupado
```

Nuevo dataframe para agrupar datos y ver como varia la eficiencia de las clases de filtro segun la hora

```python
dfAgrupadoHora = df.groupby(["filter_class", "hour"]).agg(
    eficiencia_media = ("efficiency", "mean")
).round(4).reset_index()

dfAgrupadoHora
```

Ordeno por eficiencia media, para ver horas mas eficientes

```python
dfAgrupadoHora.sort_values("eficiencia_media", ascending=False)
```

Clase Filtro con mayor eficiencia media. Ordenamos para mostrar datos
