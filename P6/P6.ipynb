```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import DataFrame
from pandas.io.parsers import TextFileReader

df: TextFileReader | DataFrame = pd.read_csv("2_air_filter.csv", encoding='utf-8-sig')
```

Creamos grafica. filter_age_days , efficiency , filter_class , location

```python
grafica1 = sns.lmplot(
    data    = df,
    x       = "filter_age_days",
    y       = "efficiency",
    hue     = "filter_class",
    col     = "location",
    scatter_kws = {"alpha": 0.3, "s": 10},
    height  = 4, # Altura en pulgadas
    aspect  = 1 #Ancho igual a altura
)

grafica1.set_axis_labels("Edad del filtro (dias)", "Eficiencia")
grafica1.set_titles("{col_name}")

plt.savefig("grafica.png", dpi=300, bbox_inches="tight")
plt.show()
```

Las graficas muestran una tendencia general y clara sobre el deterioro de la eficiencia a medida que aumenta la edad del filtro e independientemente de la localizacion y la clase de filtro. Esto es coherente con el comportamiento esperado en materiales filtrantes, donde el uso continuado reduce progresivamente su capacidad de filtrado y se requiere mantenimiento periodico

Los filtros de clase HEPA y Activated Carbon muestran una eficiencia superior al resto. Esto sugiere que estas tecnologias son mas adecuadas donde se requiera un alto nivel de filtrado de particulas.

El comportamiento es muy similar en todas las localizaciones, lo que indica que la degradacion depende principalmente de la edad del filtro y de su clase y no de su localizacion.

A partir de los 20 dias se recomienda revisar y reemplazar todos los filtros, aunque los de la clase "Activated Carbon" muestran una durabilidad mayor (+40 dias) y se puede demorar su reemplazo, su eficiencia tambien se resiente con el paso de los dias. Esta informacion permite planificar un mantenimiento de una forma mas eficiente.
